```python
##F090W

indexf090w= []
xposf090w = []
yposf090w = []
raf090w = []
decf090w = []
flux5f090w = []
flux8f090w = []
loc090w = []
with open('det_200w/f090w_with_f200w_glass.cat') as f: #input is catalogue from SE dual image more
                                                        #f090w=measure image, f200=detection image
    for _ in range(22):
        next(f)
    for line in f:
        test = re.sub("\s+", ",", line.strip()) #takes any spaces and replaces with comma to make split easier
        tokens = test.split(',')
        
        indexf090w.append(tokens[0])
        xposf090w.append(tokens[1])
        yposf090w.append(tokens[2])
        raf090w.append(tokens[3])
        decf090w.append(tokens[4])
        flux5f090w.append(tokens[5])
        flux8f090w.append(tokens[6])
f.close()


file = open('local_depths/local_depth_f090w', "r")
x_coorf090w = []
y_coorf090w = []
errorf090w = []
fixf090w = []

local_depthf090w = []
for line in file:
    test = re.sub("\s+", ",", line.strip())
    tokens = test.split(',')
    x_coorf090w.append(tokens[0])
    y_coorf090w.append(tokens[1])
    errorf090w.append(tokens[2])

for i in range(len(indexf090w)):
    x_object = xposf090w[i]#source coordinates
    y_object  =yposf090w[i]
  
    for j in range(len(x_coorf090w)): #assigns local depth from catalogue to each source if it is within the 200x200 cell
        if (float(x_coorf090w[j])-100)<=float(x_object)<=(float(x_coorf090w[j])+100):
            if (float(y_coorf090w[j])-100)<=float(y_object)<=(float(y_coorf090w[j])+100):
     

                local_depthf090w.append(errorf090w[j])#make local depth the error




file.close()                  
```
