```python
import numpy as np
```

```python
#reading in master catalogue formed from SE output 
index = []
x = []
y= []
ra=[]
dec=[]

f090w = []
f115w = []
f150w = []
f200w = []
f277w = []
f356w = []
f444w = []

f090w_err = []
f115w_err=[]
f150w_err=[]
f200w_err=[]
f277w_err=[]
f356w_err=[]
f444w_err=[]
     
with open('det_200w_master','r') as catalogue:
    
    for _ in range(1):
        next(catalogue)
    for line in catalogue:
        
        #print(len(line))
        #test = re.sub("\s+", ",", line.strip())
        tokens = line.split(',')
       #print(tokens[7])
        index.append(tokens[0])
        x.append(tokens[1])
        y.append(tokens[2])
        ra.append(tokens[3])
        dec.append(tokens[4])
        f090w.append(tokens[5])
        f090w_err.append(tokens[6])
        f115w.append(tokens[7])
        f115w_err.append(tokens[8])
        f150w.append(tokens[9])
        f150w_err.append(tokens[10])
        f200w.append(tokens[11])
        f200w_err.append(tokens[12])
        f277w.append(tokens[13])
        f277w_err.append(tokens[14])
        f356w.append(tokens[15])
        f356w_err.append(tokens[16])
        f444w.append(tokens[17])
        f444w_err.append(tokens[18])
catalogue.close()
```

```python
#making numpy array for each list
master = np.array([[index],[x],[y],[ra],[dec],[f090w], [f090w_err],[f115w],[f115w_err], [f150w],[f150w_err],[f200w],
          [f200w_err],[f277w],[f277w_err],[f356w]
                   ,[f356w_err],[f444w],[f444w_err]])
```

```python
#ensuring that the error on the flux is never greater than the flux. If greater, set flux error == error
f090w_sig = []
f115w_sig = []
f150w_sig = []
f200w_sig = []
f277w_sig= []
f356w_sig = []
f444w_sig = []

for i in range(master.shape[2]):

        if float(master[5][0][i])< (float(master[6][0][i])):   
            f090w_sig.append(float(master[6][0][i])) 
           
          
        else:
            f090w_sig.append(float(master[5][0][i]))
            
        if float(master[7][0][i])< (float(master[8][0][i])): 
            f115w_sig.append(float(master[8][0][i]))
            
        else:
            f115w_sig.append(float(master[7][0][i]))
            
        if float(master[9][0][i])< (float(master[10][0][i])): 
            f150w_sig.append(float(master[10][0][i]))
        else:
            f150w_sig.append(float(master[9][0][i]))
            
        if float(master[11][0][i])< (float(master[12][0][i])): 
            f200w_sig.append(float(master[12][0][i]))            
        else:
            f200w_sig.append(float(master[11][0][i]))
             
        if float(master[13][0][i])< (float(master[14][0][i])): 
            f277w_sig.append(float(master[14][0][i]))                          
        else:
            f277w_sig.append(float(master[13][0][i]))     

        if float(master[15][0][i])< (float(master[16][0][i])): 
            f356w_sig.append(float(master[16][0][i]))
        else:
            f356w_sig.append(float(master[15][0][i]))
                              
        if float(master[17][0][i])< (float(master[18][0][i])):                      
            f444w_sig.append(float(master[18][0][i]))
        else:
            f444w_sig.append(float(master[17][0][i]))
            
index_sig = []
x_sig = []
y_sig= []
ra_sig=[]
dec_sig=[]
f090w_err_sig = []
f115w_err_sig=[]
f150w_err_sig=[]
f200w_err_sig=[]
f277w_err_sig=[]
f356w_err_sig=[]
f444w_err_sig=[]
cslst=[]
f090wcs=[]  
f115wcs=[]  
f150wcs=[]   
f200wcs=[]   
f277wcs=[]   
f356wcs=[]   
f444wcs=[]   
#new master array to make colour selection catalogue
mastercs = np.array([[index],[x],[y],[ra],[dec],[f090w_sig], [f090w_err],[f115w_sig],
                     [f115w_err], [f150w_sig],[f150w_err],[f200w_sig],
          [f200w_err],[f277w_sig],[f277w_err],[f356w_sig],[f356w_err],[f444w_sig],[f444w_err]])
for j in range(mastercs.shape[2]):
   

     sigma_subsetcs= np.where((float(float(mastercs[11][0][j])/float(mastercs[12][0][j]))>=5)
                                 &(float(float(mastercs[5][0][j])/float(mastercs[6][0][j]))<=2)
                                 &(float(float(mastercs[7][0][j])/float(mastercs[8][0][j]))<=2),True,False)
     cslst.append(sigma_subsetcs)
for i in range(len(cslst)):
    if cslst[i]==True:
        index_sig.append(int(mastercs[0][0][i]))
        x_sig.append(float(mastercs[1][0][i]))
        y_sig.append(float(mastercs[2][0][i]))
        ra_sig.append(float(mastercs[3][0][i]))
        dec_sig.append(float(mastercs[4][0][i]))
        f090w_err_sig.append(float(mastercs[6][0][i]))
        f115w_err_sig.append(float(mastercs[8][0][i]))
        f150w_err_sig.append(float(mastercs[10][0][i]))
        f200w_err_sig.append(float(mastercs[12][0][i]))
        f277w_err_sig.append(float(mastercs[14][0][i]))
        f356w_err_sig.append(float(mastercs[16][0][i]))
        f444w_err_sig.append(float(mastercs[18][0][i]))
        f090wcs.append(float(mastercs[5][0][i]))
        f115wcs.append(float(mastercs[7][0][i]))
        f150wcs.append(float(mastercs[9][0][i]))
        f200wcs.append(float(mastercs[11][0][i]))
        f277wcs.append(float(mastercs[13][0][i]))
        f356wcs.append(float(mastercs[15][0][i]))
        f444wcs.append(float(mastercs[17][0][i]))
```

```python
#making the catalogue for Lephare

five_sig_cutf090w = []
five_sig_cutf115w = []
five_sig_cutf150w = []
five_sig_cutf200w = []
five_sig_cutf277w = []
five_sig_cutf356w = []
five_sig_cutf444w = []

index_sig = []
x_sig = []
y_sig= []
ra_sig=[]
dec_sig=[]

f090w_sig = []
f115w_sig = []
f150w_sig = []
f200w_sig = []
f277w_sig= []
f356w_sig = []
f444w_sig = []

f090w_err_sig = []
f115w_err_sig=[]
f150w_err_sig=[]
f200w_err_sig=[]
f277w_err_sig=[]
f356w_err_sig=[]
f444w_err_sig=[]

for j in range(master.shape[2]):
   

     sigma_subsetf090w= np.where((float(float(master[11][0][j])/float(master[12][0][j]))>5)
                                 &(float(float(master[5][0][j])/float(master[6][0][j]))<2)
                                 &(float(float(master[7][0][j])/float(master[8][0][j]))<2),True,False)
     five_sig_cutf090w.append(sigma_subsetf090w)


for i in range(len(five_sig_cutf090w)):
    if five_sig_cutf090w[i]==True:
        index_sig.append(int(master[0][0][i]))
        x_sig.append(float(master[1][0][i]))
        y_sig.append(float(master[2][0][i]))
        ra_sig.append(float(master[3][0][i]))
        dec_sig.append(float(master[4][0][i]))
        f090w_sig.append(float(master[5][0][i]))
        f115w_sig.append(float(master[7][0][i]))
        f150w_sig.append(float(master[9][0][i]))
        
    
        f200w_sig.append(float(master[11][0][i]))
        f277w_sig.append(float(master[13][0][i]))
        f356w_sig.append(float(master[15][0][i]))
        f444w_sig.append(float(master[17][0][i]))

#ensuring that the flux error is never less that 5% of the flux value. If it is less, then the flux error is set to 5% flux

        if float(master[6][0][i])< (0.05*float(master[5][0][i])):   
            f090w_err_sig.append(0.05*float(master[5][0][i])) 
        else:
            f090w_err_sig.append(float(master[6][0][i]))
            
            
        if float(master[8][0][i])< (0.05*float(master[7][0][i])): 
            f115w_err_sig.append(0.05*float(master[7][0][i]))
        else:
            f115w_err_sig.append(float(master[8][0][i]))
            
        if float(master[10][0][i])< (0.05*float(master[9][0][i])): 
            f150w_err_sig.append(0.05*float(master[9][0][i]))
        else:
            f150w_err_sig.append(float(master[10][0][i]))
            
        if float(master[12][0][i])< (0.05*float(master[11][0][i])): 
            f200w_err_sig.append(0.05*float(master[11][0][i]))            
        else:
            f200w_err_sig.append(float(master[12][0][i]))
             
        if float(master[14][0][i])< (0.05*float(master[13][0][i])): 
            f277w_err_sig.append(0.05*float(master[13][0][i]))                          
        else:
            f277w_err_sig.append(float(master[14][0][i]))     

        if float(master[16][0][i])< (0.05*float(master[15][0][i])): 
            f356w_err_sig.append(0.05*float(master[15][0][i]))
        else:
            f356w_err_sig.append(float(master[16][0][i]))
                              
        if float(master[18][0][i])< (0.05*float(master[17][0][i])):                      
            f444w_err_sig.append(0.05*float(master[17][0][i]))
        else:
            f444w_err_sig.append(float(master[18][0][i]))

            
#writing catalogue to file. STILL IN JANSKY UNITS
file = open('dual_image_catalogues/det_277w_master_lephare', 'w')
#file.write('index,xpos,ypos,ra,dec,f090w_flux, flux_errf090w, f115w_flux, flux_errf115w, f150w_flux, flux_errf150w,'+
 #          'f200w_flux, flux_errf200w,f277w_flux,flux_errf277w,f356w_flux flux_errf356w, f444w_flux, flux_errf444w'+'\n')
for i in range(len(index_sig)):
    file.write(str(index_sig[i])+','+str(x_sig[i])+','+str(y_sig[i])+','+str(ra_sig[i])+','+str(dec_sig[i])
               +','+str(f090w_sig[i])+','+str(f090w_err_sig[i])+','+ str(f115w_sig[i])+','+str(f115w_err_sig[i])+','
               + str(f150w_sig[i])+','+str(f150w_err_sig[i])+','+str(f200w_sig[i])+','+ str(f200w_err_sig[i])+','
               +str(f277w_sig[i])+','+str(f277w_err_sig[i])+','+str(f356w_sig[i])+','+str(f356w_err_sig[i])+','+
                str(f444w_sig[i])+','+str(f444w_err_sig[i])+'\n')
file.close()
```

```python
##Transform catalogue into magnitudes for COLOUR SELECTION  
indexcs = []
xcs = []
ycs= []
racs=[]
deccs=[]

f090wcs = []
f115wcs = []
f150wcs = []
f200wcs = []
f277wcs = []
f356wcs = []
f444wcs = []

f090w_errcs = []
f115w_errcs=[]
f150w_errcs=[]
f200w_errcs=[]
f277w_errcs=[]
f356w_errcs=[]
f444w_errcs=[]



file = open('dual_image_catalogues/det_277w_master_cs_new','r')
for line in file:
    tokens = line.split(',')
   #print(tokens[7])
    indexcs.append(tokens[0])
    xcs.append(tokens[1])
    ycs.append(tokens[2])
    racs.append(tokens[3])
    deccs.append(tokens[4])
    f090wcs.append(tokens[5])
    f090w_errcs.append(tokens[6])
    f115wcs.append(tokens[7])
    f115w_errcs.append(tokens[8])
    f150wcs.append(tokens[9])
    f150w_errcs.append(tokens[10])
    f200wcs.append(tokens[11])
    f200w_errcs.append(tokens[12])
    f277wcs.append(tokens[13])
    f277w_errcs.append(tokens[14])
    f356wcs.append(tokens[15])
    f356w_errcs.append(tokens[16])
    f444wcs.append(tokens[17])
    f444w_errcs.append(tokens[18])
```

```python
##magnitude errors 
mag_errf090w = []
mag_errf115w = []
mag_errf150w = []
mag_errf200w = []
mag_errf277w = []
mag_errf356w = []
mag_errf444w = []

for i in range(len(f090wcs)):
    mag_errf090w.append(2.5*np.log10(1+float(f090w_errcs[i])/float(f090wcs[i])))
    mag_errf115w.append(2.5*np.log10(1+float(f115w_errcs[i])/float(f115wcs[i])))  
    mag_errf150w.append(2.5*np.log10(1+float(f150w_errcs[i])/float(f150wcs[i])))  
    mag_errf277w.append(2.5*np.log10(1+float(f277w_errcs[i])/float(f277wcs[i])))
    mag_errf200w.append(2.5*np.log10(1+float(f200w_errcs[i])/float(f200wcs[i])))
    mag_errf356w.append(2.5*np.log10(1+float(f356w_errcs[i])/float(f356wcs[i])))
    mag_errf444w.append(2.5*np.log10(1+float(f444w_errcs[i])/float(f444wcs[i])))
```

```python
#flux to magnitude
f090wmag = []
f115wmag = []
f150wmag = []
f200wmag = []
f277wmag = []
f356wmag = []
f444wmag = []

for i in range(len(f090wcs)):
    f090wmag.append(-2.5*np.log10(float(f090wcs[i])*(79/85))+23.9)
    f115wmag.append(-2.5*np.log10(float(f115wcs[i])*(81/85))+23.9)
    f150wmag.append(-2.5*np.log10(float(f150wcs[i])*(81/85))+23.9)
    f200wmag.append(-2.5*np.log10(float(f200wcs[i])*(81/85))+23.9)
    f277wmag.append(-2.5*np.log10(float(f277wcs[i])*(73/85))+23.9)
    f356wmag.append(-2.5*np.log10(float(f356wcs[i])*(73/85))+23.9)
    f444wmag.append(-2.5*np.log10(float(f444wcs[i])*(71/85))+23.9)

    
```

```python
##Transform catalogue for SED

indexsed = []
xsed = []
ysed= []
rased=[]
decsed=[]

f090wsed = []
f115wsed = []
f150wsed = []
f200wsed = []
f277wsed = []
f356wsed = []
f444wsed = []

f090w_errsed = []
f115w_errsed=[]
f150w_errsed=[]
f200w_errsed=[]
f277w_errsed=[]
f356w_errsed=[]
f444w_errsed=[]
     
with open('dual_image_catalogues/det_277w_master_lephare','r') as catalogue:
#change units from janskys to ergs/s/cm2/Hz
    for line in catalogue:
        tokens = line.split(',')
        indexsed.append(int(tokens[0]))
        xsed.append(tokens[1])
        ysed.append(tokens[2])
        rased.append(tokens[3])
        decsed.append(tokens[4])
        f090wsed.append(float(tokens[5])*10**-29)
        f090w_errsed.append(float(tokens[6])*10**-29)
        f115wsed.append(float(tokens[7])*10**-29)
        f115w_errsed.append(float(tokens[8])*10**-29)
        f150wsed.append(float(tokens[9])*10**-29)
        f150w_errsed.append(float(tokens[10])*10**-29)
        f200wsed.append(float(tokens[11])*10**-29)
        f200w_errsed.append(float(tokens[12])*10**-29)
        f277wsed.append(float(tokens[13])*10**-29)
        f277w_errsed.append(float(tokens[14])*10**-29)
        f356wsed.append(float(tokens[15])*10**-29)
        f356w_errsed.append(float(tokens[16])*10**-29)
        f444wsed.append(float(tokens[17])*10**-29)
        f444w_errsed.append(float(tokens[18])*10**-29)
catalogue.close()
```
