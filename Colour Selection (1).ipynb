```python
import numpy as np
```

```python

index = []
x = []
y= []
ra=[]
dec=[]

f090w = []
f115w = []
f150w = []
f200w = []
f277w = []
f356w = []
f444w = []

f090w_err = []
f115w_err=[]
f150w_err=[]
f200w_err=[]
f277w_err=[]
f356w_err=[]
f444w_err=[]
     
with open('dual_image_catalogues/det_200w_master_mags_cs_new','r') as catalogue:

    for line in catalogue:
        tokens = line.split(',')
        index.append(tokens[0])
        x.append(tokens[1])
        y.append(tokens[2])
        ra.append(tokens[3])
        dec.append(tokens[4])
        f090w.append(tokens[5])
        f090w_err.append(tokens[6])
        f115w.append(tokens[7])
        f115w_err.append(tokens[8])
        f150w.append(tokens[9])
        f150w_err.append(tokens[10])
        f200w.append(tokens[11])
        f200w_err.append(tokens[12])
        f277w.append(tokens[13])
        f277w_err.append(tokens[14])
        f356w.append(tokens[15])
        f356w_err.append(tokens[16])
        f444w.append(tokens[17])
        f444w_err.append(tokens[18])  
        
    
master = np.array([[index],[x],[y],[ra],[dec],[f090w], [f090w_err],[f115w],[f115w_err], [f150w],[f150w_err],[f200w],
          [f200w_err],[f277w],[f277w_err],[f356w],[f356w_err],[f444w],[f444w_err]])

catalogue.close()
```

```python
#HARIKANE METHOD - z=9 
test = []
z9_index = []
for i in range(master.shape[2]):
    subset = np.where((float((master[7][0][i]))-float(master[9][0][i])>1)
              & ((float(master[9][0][i])-float(master[13][0][i]))<1)
              & ((float(master[7][0][i])-float(master[9][0][i])>(float(master[9][0][i])-float(master[13][0][i])+1)))
              ,True, False)
    test.append(subset)
    
for i in range(len(test)):
    if test[i]==True:
        z9_index.append(int(master[0][0][i]))
cs_catalogue_indexes=[]
for j in range(len(x)):
    for k in z9_index:
        if int(index[j])==int(k):
            cs_catalogue_indexes.append(j)
```
